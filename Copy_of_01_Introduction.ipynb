```python
target = 'Species'
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalWidthCm', 'PetalLengthCm']
```


### Building Your Model
You will use the scikit-learn library to create your models. Scikit-learn is the most popular library for modeling the types of data typically stored in DataFrames. It contains so many different machine learning algorithms

The steps to building and using a model are:

Define: What type of model will it be? A linear regression?, A decision tree?
Fit: Capture patterns from provided data. This is the heart of modeling.
Predict: Just what it sounds like
Evaluate: Determine how accurate the model's predictions are.

In this example, we will choose a decision tree

```python
from sklearn.tree import DecisionTreeClassifier

# Define model. Specify a number for random_state to ensure same results each run
model = DecisionTreeClassifier(random_state=1)

# Fit model
model.fit(dataframe[features], dataframe[target])
```

```python
dataframe[features].head()
```

```python
print("Making predictions for the following 5 flowers")
print(dataframe[features].head())
print("The predictions are")
print(model.predict(dataframe[features].head()))
```

### Evaluating the model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(dataframe[features], dataframe[target], random_state = 0)
# Define model
model = DecisionTreeClassifier()
# Fit model
model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = model.predict(val_X)
print(accuracy_score(val_y, val_predictions))
```

### Underfitting and Overfitting

Overfitting: capturing spurious patterns that won't recur in the future, leading to less accurate predictions, or

Underfitting: failing to capture relevant patterns, again leading to less accurate predictions.
We use validation data, which isn't used in model training, to measure a candidate model's accuracy. This lets us try many candidate models and keep the best one.

```python
def get_score(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    score = accuracy_score(val_y, preds_val)
    return(score)
```

```python
# compare accuracy score with differing values of max_leaf_nodes
for max_leaf_nodes in [2, 3, 5, 50, 500]:
    my_score = get_score(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes}  \t\t Accuracy Score:  {my_score}")
```

Max Leaf node of 5 is the lowest leaf node with the best accuracy

### Saving and Loading the model

```python
model = DecisionTreeClassifier(max_leaf_nodes=5, random_state=0)
model.fit(train_X, train_y)

### Save the model
import pickle

# Save the model to disk
with open('tree_model.pkl', 'wb') as file:
  pickle.dump(model, file)

# Loading the model from disk
with open('tree_model.pkl', 'rb') as file:
  treeModel = pickle.load(file)

# Use the loaded model to make predictions on new data
# predictions = treeModel.predict(...)
```
