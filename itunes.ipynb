```python
import json
import requests
import sys

# Check if a search term was provided via command line
if len(sys.argv) == 2:
    search_term = sys.argv[1]
else:
    # If no command-line argument, prompt for input
    search_term = input("Enter search term: ")
    
# Make the API request
response = requests.get("https://itunes.apple.com/search?entity=song&limit=50&term=" + search_term)

# Parse the response
o = response.json()
if "results" in o and len(o["results"]) > 0:
    for result in o["results"]:
        print(result["trackName"])
else:
    print("No results found")

# Print the full JSON response
print(response.json())
```
